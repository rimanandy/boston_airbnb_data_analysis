```python
#Wrangling and cleaning data
# choosing columns which are important to us in this project
cols = ['id', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 'property_type', 'accommodates', 'bathrooms', 'beds', 'price', 'review_scores_value', 'review_scores_rating', 'review_scores_cleanliness', 'review_scores_communication', 'review_scores_location', 'neighbourhood']
```

```python
listings = listings[cols]
listings.head()
```

```python
wrangle_prices = lambda x : (x.replace('.00','').replace('$','').replace(',',''))
listings['price']=listings['price'].apply(wrangle_prices)
listings['price'] = pd.to_numeric(listings['price'])
# converting the prices to numeric value
listings.head()
```

```python
# removing % sign and converting host_response_rate to numeric
wrangle_data = lambda x : (x.replace('%',''))
listings['host_response_rate']=listings['host_response_rate'].apply(wrangle_data)
listings['host_response_rate'] = pd.to_numeric(listings['host_response_rate'])
listings['host_response_rate'].head()
```

```python
#Q1. How well are property and prices related?
```

```python
features_df = listings.copy()
```

```python
# checking if the target label has all entries recorded
features_df.price.count()
```

```python
features_df.shape
```

```python
# categorizing number of listings with property type
property_type_count = features_df.property_type.value_counts()
```

```python
property_type_count
```

```python
property_type_count.plot.bar(color = 'teal', figsize =(10,5), title = 'Boston Property Type Frequency')
plt.xlabel('Property type');
plt.ylabel('Number of listings');
```

```python
price_per_property_type = features_df.groupby('property_type').price.mean()
```

```python
price_per_property_type
```

```python
# plot a graph that depicts a relation between property type and price
plt.figure(figsize=(10,5));
price_per_property_type.plot(kind='bar', color='teal');
plt.xlabel('Property Types')
plt.ylabel('Average Price')
plt.title('Property vs Average Price Chart');
```

```python
#Conclusions:
#Airbnb has more apartments listings followed by Bed and Breakfast.
#The property type is largely dependent on the price.
# Guesthouse property type has the highest price followed by Boat and Loft.
```

```python
#Q2. Where to invest in a property in Boston to get the maximum number of returns from Airbnb?
```

```python
price_per_neighbourhood = features_df.groupby('neighbourhood').price.mean()
```

```python
price_per_neighbourhood
```

```python
# plot a graph that depicts a relation between neighbourhood and price
plt.figure(figsize=(15,8));
price_per_neighbourhood.plot(kind='bar', color='teal');
plt.xlabel('Neighbourhoods')
plt.ylabel('Average Price')
plt.title('Neighbourhood vs Average Price Chart');
```

```python
#Conclusions:
#Investing a property in the financial district will provide the highest returns followed by Downtown Crossing and Leather District.
```

```python
#How well can we predict reviews and what aspects correlate with reviews?
```

```python
features_df.review_scores_rating.unique()
```

```python
ratings_per_neighbourhood = features_df.groupby(
    ['neighbourhood', 'property_type']).review_scores_rating.mean()
```

```python
ratings_per_neighbourhood.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='neighbourhood', y='property_type', hue='review_scores_rating', data=features_df);
plt.xticks(rotation=90);
plt.title('Reviews based on neighbourhood and price');
```

```python
ratings_per_property_type = features_df.groupby('property_type').review_scores_rating.mean()
ratings_per_property_type
```

```python
ratings_per_property_type.plot(kind='bar',color='teal');
plt.xlabel('Property types')
plt.ylabel('Average rating score')
plt.title('Property Type vs Average Rating Chart');
```

```python
ratings_per_bedrooms = features_df.groupby(['property_type', 'beds']).review_scores_rating.mean()
```

```python
ratings_per_bedrooms
```

```python
plt.figure(figsize=(15, 5))
ratings_per_bedrooms.plot(kind='bar',color='teal');
plt.xlabel('Property type with number of bedrooms')
plt.ylabel('Average Rating Score')
plt.title('Property_type, Bedrooms vs Average Rating Chart');
```

```python
ratings_per_bathrooms = features_df.groupby(['property_type', 'bathrooms']).review_scores_rating.mean()
ratings_per_bathrooms
```

```python
plt.figure(figsize=(15, 5))
ratings_per_bathrooms.plot(kind='bar',color='teal');
plt.xlabel('Property type with number of bathrooms')
plt.ylabel('Average Rating Score')
plt.title('Property_type, Bathrooms vs Average Rating Chart');
```

```python
#Conclusion
#From the data it is obvious that people enjoy living in Guesthouses, Entire Floor and Boats  depending on their interests and budgets.

#There seems to be Guesthouse with only 2 bedrooms and 1 bathroom in listings data and these are likely to get the highest review score. People are also comfortable with Villas with 2 bathrooms and 2 bedrooms . 
#However, they are not quite satisfied with Bed and Breakfast's experiences as they have got the least review score.
#Again, guesthouse takes the first place along with entire floow with at least one bathroom and villas with two.
```
